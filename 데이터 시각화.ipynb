```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('04_data1.xlsx', index_col = 'area')
data.head()
```

```python
data2020_good = data['2020_good']
```

```python
data2019_good = data['2019_good']
```

```python
plt.figure(figsize=(20, 4))
plt.plot(data2020_good, label = 2020, color = 'b', marker = 'o')
plt.plot(data2019_good, label = 2019, color = 'r', marker = 's')
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2020 vs 2019')
plt.grid(True)# 
plt.legend()
plt.show()
```

```python
data2 = pd.read_csv('04_data2.csv')
data2
```

```python
PetalLength = data2['PetalLength']
PetalWidth = data2['PetalWidth']

plt.figure(figsize = (10, 4))
plt.scatter(PetalLength, PetalWidth)

plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Iris Petal Scatter Graph')
plt.grid
plt.show
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
newdata1 = df['Sex']
newdata2 = df['Age']

new_df = pd.concat([newdata1, newdata2], axis = 1)
```

```python
sex = new_df['Sex']
age = new_df['Age']
plt.figure(figsize=(10, 4))
plt.scatter(sex, age, color = 'r')
plt.xlabel('sex')
plt.ylabel('age')
plt.title('survivors')
plt.grid
plt.show
#    ,     
```
