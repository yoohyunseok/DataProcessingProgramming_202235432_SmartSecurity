<h2>4 번</h2>
Violinplot (지름/높이)

```python
diameter = df['Diameter']
height = df['Height']
```

```python
data = pd.concat([diameter, height], axis = 1)
```

```python
plt.figure(figsize = (8, 5))
sns.violinplot(data = data)

plt.title('Height & Weight')
plt.legend
plt.show
```
