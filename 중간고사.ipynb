```python
# 승객이 0인 픽업 위도경도 로 구성된 zeropickup 생성
condition = (df['passenger_count'] == 0)
zeropickup = df.loc[condition, ['pickup_longitude', 'pickup_latitude']]
zeropickup.head()
```

```python
# 승객이 0인 드롭 위도경도 로 구성된 zeropickup 생성
zerodropoff = df.loc[condition, ['dropoff_longitude', 'dropoff_latitude']]
zerodropoff.head()
```

```python
# 픽업과 드롭에 따른 위도경도 산점도 작성
pickup_longtitude = zeropickup['pickup_longitude']
pickup_latitude = zeropickup['pickup_latitude']

dropoff_longtitude = zerodropoff['dropoff_longitude']
dropoff_latitude = zerodropoff['dropoff_latitude']

plt.figure(figsize = (10,4))
plt.scatter(pickup_longtitude, pickup_latitude, color = 'r', edgecolor = 'w', label = 'Pickup Location')
plt.scatter(dropoff_longtitude, dropoff_latitude, color = 'b', edgecolor = 'w', label = 'Dropoff Location')

plt.xlabel('Longtitude')
plt.ylabel('Latitude')
plt.legend()
plt.grid()
plt.show
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data6['weekday']=le.fit_transform(data6['weekday'])
```

```python
# 1월과 5월에 승객수와 요일
condition_january = (data6['month'] == 1)
condition_may = (data6['month']==5)
hist1 = data6.loc[condition_january, ['passenger_count']]
hist2 = data6.loc[condition_may, ['passenger_count']]
```

```python
hist1.head()
```

```python
hist2.head()
```

```python
plt.figure(figsize = (10, 6))
plt.hist(hist1, alpha = 0.5, bins = 5, label = 'Jan')
plt.hist(hist2, bins = 5, alpha = 0.5, label = 'Fab')
plt.xlabel('weekday')
plt.ylabel('Passenger Count')
plt.title('Passenger count per weekday')
plt.legend()
plt.grid()
plt.show
```

```python
# 동일한 데이터로  KDE 히스토그램 작성
plt.figure(figsize = (10, 6))
sns.histplot(hist1, alpha = 0.5, bins = 5, label = 'Jan', kde = True)
sns.histplot(hist2, bins = 5, alpha = 0.5, label = 'Fab', color = 'r', kde = True)
plt.xlabel('weekday')
plt.ylabel('Passenger Count')
plt.title('Passenger count per weekday')
plt.legend()
plt.grid()
plt.show
```
