```python
df3 = pd.read_excel("C:/Users/UserName/Desktop/2-2학기/스보/데이터 처리 프로그래밍/05_data1.xlsx")
```

```python
data6 = df3.loc[:, ['gender', 'height', 'weight', 'waist', 'drinking', 'smoking']]
data6.loc[data6['gender']==1, ['gender']]='M'
data6.loc[data6['gender']==0, ['gender']]='F'
data6.loc[data6['drinking']==0, ['drinking']]='Non-drinking'
data6.loc[data6['drinking']==1, ['drinking']]='Drinking'
data6.loc[data6['smoking']==1, ['smoking']]='Non-smoking'
data6.loc[data6['smoking']==2, ['smoking']]='Quit-smoking'
data6.loc[data6['smoking']==3, ['smoking']]='Smoking'
```

```python
drinking = data6.groupby(['gender', 'drinking'])['drinking'].count()
smoking = data6.groupby(['gender', 'smoking'])['smoking'].count()
drinking = drinking.to_frame(name='count')
smoking = smoking.to_frame(name = 'count')
drinking = drinking.reset_index()
smoking = smoking.reset_index()
```

```python
fig = plt.figure(figsize = (12,10))
fig.suptitle('2020 Health Check Drinking & Smoking Type Bar Graph', fontweight = 'bold')
index1 = np.arange(4)
index2 = np.arange(6)

fig.add_subplot(2,1,1)

plt.bar(index1, drinking['count'])
plt.title('Drinking Type')
plt.ylabel('Count')
plt.xticks(index1, ['Non-drinking(M)', 'Drinking(M)', 'Non-drinking(F)', 'Drinking(F)'])

fig.add_subplot(2, 1, 2)
plt.bar(index2, smoking['count'])
plt.title('Smoking Type')
plt.ylabel('Count')
plt.xticks(index2, ['Non-smoking(M)', 'Quit-smoking(M)', 'Smoking(M)', 'Non-snoking(F)', 'Quit-smoking(F)', 'Smoking(F)'])
plt.show()
```

```python
fig = plt.figure(figsize = (12,12))

area1 = fig.add_subplot(2,1,1)
area2 = fig.add_subplot(2, 1, 2)
ax1 = sns.barplot(x='gender', y='count', hue='drinking', data = drinking, ax = area1)
ax2 = sns.barplot(x = 'gender', y='count', hue='smoking', data=smoking, ax = area2)

fig.suptitle('2020 Health Check Drinking & Smoking Type Bar Graph', fontweight = 'bold')
area1.set_title('Drinking Type')
area2.set_title('Smoking Type')
plt.show()
```

```python
#age, height, weight, blood_sugar만 가져와서 새로운 new df로 만든다.
#gender 나눔
#blood_Sugar가 120 넘는 사람과 안넘는 사람 구분->그룹화해서 count
#해서 seaborn 막대그래프 그리기. x축 gender, y축 수
```

```python
new_df = df3.loc[:, ['gender', 'height', 'weight', 'blood_sugar']].head(100)
new_df
```

```python
BMI = new_df['weight']/(new_df['height']*new_df['height'])
```

```python
BMI
```

```python
new_df['BMI'] = BMI*10000
```

```python
new_df
```

```python
new_df.loc[new_df['gender']==1, ['gender']]='M'
new_df.loc[new_df['gender']==2, ['gender']]='F'
```

```python
condition_bloodsugar_high = (new_df['blood_sugar']>120)
condition_bloodsugar_low = (new_df['blood_sugar']<=120)
```

```python
new_df.loc[condition_bloodsugar_high, ['blood_sugar']] = '120_above'
new_df.loc[condition_bloodsugar_low, ['blood_sugar']] = '120_under'
```

```python
condition_bmi_high = (new_df['BMI']>25)
condition_bmi_low = (new_df['BMI']<23)
condition_bmi_middle = (new_df['BMI']>=23)&(new_df['BMI']<=25)
```

```python
new_df.loc[condition_bmi_high, ['BMI']] = 'obesity'
```

```python
new_df.loc[condition_bmi_low, ['BMI']] = 'normal'
```

```python
new_df.loc[condition_bmi_middle, ['BMI']] = 'overweight'
```

```python
new_df
```

```python
blood_sugar = new_df.groupby(['gender', 'blood_sugar'])['blood_sugar'].count()
bmi = new_df.groupby(['gender', 'BMI'])['BMI'].count()
blood_sugar = blood_sugar.to_frame(name = 'count')
bmi = bmi.to_frame(name = 'count')
blood_sugar = blood_sugar.reset_index()
bmi = bmi.reset_index()
```

```python
fig = plt.figure(figsize = (12, 12))

area1 = fig.add_subplot(2,1,1)
area2 = fig.add_subplot(2,1,2)
ax1 = sns.barplot(x = 'gender', y='count', hue = 'blood_sugar', data = blood_sugar, ax = area1)
ax2 = sns.barplot(x = 'gender', y='count', hue = 'BMI', data = bmi, ax = area2)

fig.suptitle('report')
area1.set_title('blood sugar figure')
area2.set_title('bmi figure')

plt.show
```
